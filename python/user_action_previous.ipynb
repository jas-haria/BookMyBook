```python
import pandas as pd
import numpy as np
from flask import Flask, request, jsonify
from flask_cors import CORS
```

```python
app = Flask(__name__)
CORS(app)
```

```python
@app.route('/',methods=['POST'])
def action():
    #read csv and process dataframe
    data = pd.read_csv("C:/Users/Saloni/Desktop/Dataset/newsrecomm2new.csv")
    if 'Unnamed: 0' in data.columns:
        data = data.drop('Unnamed: 0', axis=1)
    data.set_index('url',inplace=True)
    #get data from post request
    userData = request.get_json(force=True)
    rating='99'
    if(userData['url'] in data.index):
        if (data.loc[userData['url'],userData['email']]=='0'):
            rating = '-10'
        if (data.loc[userData['url'],userData['email']]=='10'):
            rating = '10'
    return jsonify({'rating' : rating})
```

```python
if __name__ == '__main__':
    app.run(port=5001, debug=False)
```
